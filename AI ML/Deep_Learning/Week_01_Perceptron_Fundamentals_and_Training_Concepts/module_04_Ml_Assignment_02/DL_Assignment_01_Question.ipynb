# Question 01: [ Marks 10 ]

What is a perceptron?
Explain its three main components and their roles.

**Write** Answer 01:
A Perceptron is the simplest type of artificial neural network. It is used to classify data into two classes by learning from input data.

**Three Main Components of a Perceptron**
1. Input (x)
  - Receives the input features or data.
  - Example `x1`, `x2`, `x3`
2. Weight(w)
  - Each input has weight
  - Weigts decide how much each input influences the final output.
3. Activation Function
  - Calculates the weighted sum and produces the final output.
  - If the value is greater than 0, the output is 1, otherwise it is 0

Why is the perceptron called a linear classifier?
Explain your answer using the concept of linear separability and logical gate examples such as AND, OR, and XOR.

## Write Answer 03:
A perceptron is called a linear classifier because it separates data using a strainght line (or a hyperplance in higher dimensions). It can correctly classify only linearly separabel data.
Linear separability means that two classes can be separated by a single straight line.

Examples:
  - AND gate: Linearly separable so a perceptron can classify it correctly.
  - OR gate: Also linearly separable, so a perceptron works well.
  - XOR gate: Not linearly separable because no single straight line can separate its outputs. Therefore a single perceptron cannot solve the XOR problem.

# Question 04: [ Marks 15 ]

Write the perceptron weight update rule and explain it intuitively.
Why are the weights updated only when the prediction is incorrect?

## Write Answer 04:
The perceptron weight update rule is:
w(new) = w(old) + η * ( y - ŷ) * x

- η = Learning rate
- y = Actual output
- ŷ = Predicted output
- x = Input value

**Explanation:**
The perceptron compares the predicted output with the actual output. If the prediction is wrong, it updates the weights to reduce the error. This helps the model make a better prediction next time.

### Why are weights updated only when the prediction is incorrect?
Weights are updated only when the prediction is incorrect because the model needs to learn from its mistakes. If the prediction is already correct, there is no error, so the currect weights are working well and do not need to be changed. This makes the learning process simple and efficient.

# Question 06: [ Marks 20 ]

What is meant by the “perceptron Learning Rule” or weight adjustment method?
Why can adjusting one weight affect the classification of other points?

## Write Answer 06:
The Perceptron Learning Rule is the method used to update the weights when the perceptron makes a wrong prediction. Its goal is to reduce the error and improve futher predictions.

The weight update rule is: w(new) = w(old) + η * (y - ŷ) * x

- η = Learning rate
- y = Actual output
- ŷ = Predicted output
- x = Input value

### Why can adjusting one weight affect the classification of other points?
A single weight is shared by all input data. When one weight is updated, the decision boundary changes. As a result some other data points may move to a different side of the boundary so their classification can also change.

This is why changing one weight can improve the prediction for one point but may also affect the classification of other points.

### Part B – Percentiles, IQR and Z-Score

```python
weight = [2.5, 3.0, 3.2, 3.3, 3.4, 3.5, 3.6, 3.9, 4.0, 4.5]
lower_half = weight[:5]
q1 = (lower_half[1] + lower_half[2]) /2
q1 = lower_half[2]
print(f"Q1 = {q1}")

q2 = (weight[4] + weight[5]) / 2
print(f"Q2 {q2}")
upper_half = weight[5:]
q3 = upper_half[2]
print(f"Q3 {q3}")

IQR = q3 - q1
print(f"IQR {IQR}")
lower_fence = q1 - 1.5 * IQR
print(f"lower fence {lower_fence}")
upper_fence = q3 - 1.5 * IQR
print(f"upper fence {upper_fence}")

outliers = [x for x in weight if x < lower_fence or x > upper_fence]
print(f"outliers {outliers if outliers else 'None'}")
```

2. Email dataset of 1 000 mails:


```python
total = 1000
spam = 100
free_spam = 40
free_nonspam = 60

ps = spam / total
pNots = 1 - ps

pf_s = free_spam / spam
pf_nots = free_nonspam / (total - spam)

pf = pf_s * ps + pf_nots * pNots
ps_give_f = (pf_s * ps)/ pf
print(f"P(S|F) = {ps_give_f:.4f}")
```

```python
pd = 1 / 10000
pPlus_d = 0.99
pNeg_d = 0.99
pPos_nod = 1-pNeg_d

ppos = pPlus_d * pd + pPos_nod * (1-pd)
pd_give_pos = (pPlus_d* pd)/ ppos

print(f"P(D|+) = {pd_give_pos:.6f}")
print(f"{pd_give_pos * 100:.2f}%")
```

```python
# Hospital Triage — All Tasks
# Multiclass Confusion Matrix (U / N / R)

cm = {
    'U': {'U': 120, 'N': 42,  'R': 18},
    'N': {'U': 60,  'N': 612, 'R': 48},
    'R': {'U': 45,  'N': 90,  'R': 165},
}
classes = ['U', 'N', 'R']
total = 1200

print(f"{'Class':<6} {'Prec':>7} {'Recall':>8} {'F1':>7} {'NPV':>7}")
for cls in classes:
    tp = cm[cls][cls]
    fp = sum(cm[other][cls] for other in classes if other != cls)
    fn = sum(cm[cls][other] for other in classes if other != cls)
    
    tn = total - tp - fp - fn 
    prec = tp / (tp + fp)
    recall = tp / (tp + fn)
    f1 = 2* prec * recall / (prec + recall)
    npv = tn / (tn + fn)
    print(f"{cls:<6} {prec:>7.3f} {recall:>8.3f} {f1:>7.3f} {npv:>7.3f}")
```

```python
correct = sum(cm[c][c] for c in classes)
accuracy = correct / total
print(f"\nTask 2 — Overall Accuracy = {correct}/{total} = {accuracy:.4f}")
```

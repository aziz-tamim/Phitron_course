```python
# B1. Three scalers side by side

# Given:
# Heights = [150,160,170,175,180]
# Weights = [58,62,65,66,190]

# a) Min-Max scale both to [0,1]

# Formula:
# x' = (x-min)/(max-min)

# Heights:
# min =150
# max =180
# range =30

# 150 = (150-150)/30 = 0
# 160 = (160-150)/30 = 10/30 = 0.333
# 170 = (170-150)/30 = 20/30 = 0.667
# 175 = (175-150)/30 = 25/30 = 0.833
# 180 = (180-150)/30 = 30/30 = 1

# Final:
# [0,0.333,0.667,0.833,1]

# Weights:
# min =58
# max =190
# range =132

# 58 = (58-58)/132 =0
# 62 = (62-58)/132 =4/132 =0.030
# 65 = (65-58)/132 =7/132 =0.053
# 66 = (66-58)/132 =8/132 =0.061
# 190=(190-58)/132 =1

# Final:
# [0,0.030,0.053,0.061,1]

# b) Standardize first three values only

# Heights = [150,160,170]

# Mean=(150+160+170)/3
# Mean=160

# SD=8.16

# Formula:
# z=(x-mean)/SD

# 150=(150-160)/8.16=-1.22
# 160=(160-160)/8.16=0
# 170=(170-160)/8.16=1.22

# Final:
# [-1.22,0,1.22]

# Weights=[58,62,65]

# Mean=61.67
# SD=2.87

# 58=(58-61.67)/2.87=-1.28
# 62=(62-61.67)/2.87=0.12
# 65=(65-61.67)/2.87=1.16

# Final:
# [-1.28,0.12,1.16]

# c) Robust scale Weights

# Weights=[58,62,65,66,190]

# Median=65

# Q1=(58+62)/2
# Q1=60

# Q3=(66+190)/2
# Q3=128

# IQR=Q3-Q1
# IQR=68

# Formula:
# x'=(x-Median)/IQR

# 58=(58-65)/68=-0.10
# 62=(62-65)/68=-0.04
# 65=(65-65)/68=0
# 66=(66-65)/68=0.015
# 190=(190-65)/68=1.84

# Final:
# [-0.10,-0.04,0,0.015,1.84]

# d) Which scaler handles outlier best?

# Robust Scaling handles outliers best
# because it uses Median and IQR.
```

```python
n = int(input())
arr = list(map(int,input().split()))
mn_val = min(arr)
mx_val = max(arr)

mn_idx = arr.index(mn_val)
mx_idx = arr.index(mx_val)

arr[mn_idx], arr[mx_idx] = arr[mx_idx], arr[mn_idx]

for x in arr:
    print(x, end=" ")
```

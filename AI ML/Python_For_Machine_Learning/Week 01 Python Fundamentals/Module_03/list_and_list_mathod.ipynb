### List as Stack

```python
stack = []

# stack e push
stack.append(1)
stack.append(2)
stack.append(3)
stack.append(4)
stack.append(5)

print(f"top element : {stack[-1]}")

# removing the top element
stack.pop()

# top element
print(f"top element : {stack[-1]}")


```

### List as Queue

```python
queue = []

# insertion
queue.append(1)
queue.append(2)
queue.append(3)
queue.append(4)
queue.append(5)


# remove 
print(queue.pop(0))


# accessing the front element
print(f"front e {queue[0]}")

queue.pop(0)
print(f"fornt e {queue[0]}")
```

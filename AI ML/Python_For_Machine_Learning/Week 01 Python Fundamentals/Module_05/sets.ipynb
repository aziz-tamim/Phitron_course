### Set Mathematical Methods

```python
a = {1, 2, 3}
b = {5, 6, 4}
print(a.union(b))
print(a.intersection(b))
print(a.isdisjoint(b))
print(a.issubset(b))
```

### Dictionaries in Python

```python
dic = {}
print(type(dic))

dic = {"name": "Phitron", "age": 20, "address" : "Dhaka", "numbers" : [10, 20, 30]}
print(type(dic), dic)
```

```python
# access
# dic_name[key]
print(dic["numbers"])
print(dic.get("age"))

dic["name"] = "Phitron AI/ML"
dic = {"name": "Phitron", "age": 20, "address" : "Dhaka", "numbers" : [10, 20, 30], "age":30}
print(dic)
```

```python
# printing default values
print(dic)
print(dic.get("math_marks"))
print(dic.get("math_marks",0))
```

```python
dic = {"name": "Phitron", "age": 20, "address" : "Dhaka", "numbers" : [10, 20, 30]}
print(dic)
dic["math_marks"] = 30
print(dic)

dic.update({"English_marks":40})
print(dic)
```

```python
# deleting
del dic["math_marks"]
print(dic)
```

```python
dic_2 = dic.copy()
print(dic_2)
```

```python
# testing
dic_3 = { {"name" : "adil", "age" : 20} : 20}
print(dic_3)
```

### Iterating over Dictionary

```python
dic = {"name": "Phitron", "age": 20, "address" : "Dhaka", "numbers" : [10,20,30], "age":30}
keys = dic.keys()
print(keys)

dic["math_numbers"] = 30
print(keys)

values = dic.values()
print(values)

dic["math_numbers"] = 40
print(values)

items = dic.items()
print(items)
```

```python
# iteration

for key, value in dic.items():
    print(key, value)
```

### Dictionary Comprehension

```python
square = { x:x**2 for x in range(1,11) if x%2 == 0}
print(square)
```

```python
co_ordinates = [(10,10.5), (20.5,192), (101,102)]
location = ["dhaka", "chattogram", "sylhet"]

exact_location = {co_or : loc for co_or, loc in zip(co_ordinates, location)}
print(exact_location)
```

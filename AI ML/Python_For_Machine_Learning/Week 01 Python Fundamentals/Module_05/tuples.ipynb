```python
# declaration

tup = (10, 20, 30, 40)
# print(type(tup), tup)

float_tup = (10.5, 10.2)
# print(type(float_tup), float_tup)

mixed_tup = (10, 10.5, "str", True)
# print(mixed_tup)

tup1 = ([10])
print(type(tup1), tup1)
```

```python
# Access
tup = (10, 20, 30, 10.5, "str", False)

# index
print(tup[3])

#slicing
new_tup = tup[0:3]
print(new_tup)
```

```python
## immutability

list = [ 10, 20, 30, 40 ]
list.append(60)
list[1] = 100
print(list)

tup = (10, 20, 30, 40)
tup[1] = 100
print(tup)
```

```python
# tuple method
tup = (10, 20, 30, 40, 10, 20 , 30, 10)
print(tup.count(10))
print(tup.index(40))
```

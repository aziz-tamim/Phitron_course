### function parameters ( Input )

```python
def greet(user):
    print(f"hello {user}")
```

```python
def square_addition(a,b,c):
    print(f"{a} {b} {c}")
    a = a**2
    b = b**2
    return a+b+c
    
# positional argument
ans = square_addition(10,20,30)+10
print(ans)

#keyword argument
ans = square_addition(b=10,c=20,a = 30)+10
print(ans)
```

```python
def square_addition(*args):
    summation = 0
    for i in args:
        i = i**2
        summation += i
    return summation
```

```python
ans = square_addition(10,20,30,40,50)
print(ans)
```

```python
def student(**kwargs):
    for key, val in kwargs.items():
        print(f"{key} : {val}")
```

```python
student(name = "tamim", cls = 10, roll = 30, marks = 20.5)
```

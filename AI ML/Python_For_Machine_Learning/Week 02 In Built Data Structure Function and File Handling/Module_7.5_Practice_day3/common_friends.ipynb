### Problem 1: Common Friends (Set Operations)
###### Problem:
 Given two sets of friends from two people, find the mutual friends, unique friends of each, and the total number of unique friends.
###### Example Input:
###### a_friends = {"Rahim", "Karim", "Sakib", "Jamal"}
###### b_friends = {"Sakib", "Jamal", "Rafiq", "Nadim"}

###### Expected Output:
###### Mutual friends: {'Sakib', 'Jamal'}
###### Unique to A: {'Rahim', 'Karim'}
###### Unique to B: {'Rafiq', 'Nadim'}
###### Total unique friends: 6



```python
friends_a = {"Ramim", "Karim", "Sakib", "Jamal"}
friends_b = {"Sakib", "Jamal", "Rafiq", "Nadim"}

mutual = friends_a & friends_b

unique_A = friends_a - friends_b
unique_B = friends_b - friends_a

total_unique_friend_cnt = len(friends_a | friends_b)

print(f"Mutal friends: {mutual}\nUnique to A: {unique_A}\nUnique to B: {unique_B}\nTotal unique friends: {total_unique_friend_cnt}")
```

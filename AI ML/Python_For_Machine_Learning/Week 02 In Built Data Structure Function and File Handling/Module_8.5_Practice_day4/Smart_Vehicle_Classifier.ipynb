```python
class Vehicle:
    def __init__(self, vehicle_type, name, speed):
        self.vehicle_type = vehicle_type
        self.name = name
        self.__speed = speed
    
    def get_speed(self):
        return self.__speed
        
    def classify(self):
        if self.__speed > 100:
            return "High-speed Vehicle"
        else:
            return "Normal Vehicle"
    
class Car(Vehicle):
    def show_info(self):
        print(f"Type: {self.vehicle_type}, Name: {self.name}, Speed: {self.get_speed()} km/h, Predicted: {self.classify()}")
        
class Bike(Vehicle):
    def show_info(self):
        print(f"Type: {self.vehicle_type}, Name: {self.name}, Speed: {self.get_speed()} km/h, Predicted: {self.classify()}")


inp = int(input())
Vehicle = []

for i in range(inp):
    vehicle_type, name, speed = input().split()
    speed = int(speed)
    
    if vehicle_type == "Car":
        car_obj = Car(vehicle_type, name, speed)
        print("Car Added:", name,)
    else:
        car_obj = Bike(vehicle_type, name, speed)
        print("Bike Added:", name,)
    Vehicle.append(car_obj)
    
print("\n--- Vehicle Details ---")
for x in Vehicle:
    x.show_info()
```

## 1 . Create a 1D, 2D, and 3D NumPy array and print their dimensions and shapes.


```python
import numpy as np

# 1D array
arr1 = np.array([1,2,3,4,5])

print("1D Dimention",arr1.ndim)
print("1D shape", arr1.shape)

#2D array
arr2 = np.array([[1,2,3],
                 [4,5,6]
                ])

print("2D Dimention",arr2.ndim)
print("2D shape", arr2.shape)

#3D array
arr3 = np.array([
    [
        [1,2,3],
        [4,5,6]
    ],
    [
        [1,2,3],
        [4,5,6]
    ],
    [
        [1,2,3],
        [4,5,6]
    ]
])

print("3D Dimention",arr3.ndim)
print("3D shape", arr3.shape)
```

```python
zero_shape_arr = np.zeros((3,3))
print(zero_shape_arr)

ones_shape_arr = np.ones((2,5))
print(ones_shape_arr)

fixed_arr = np.full((3,3),7)
print(fixed_arr)
```

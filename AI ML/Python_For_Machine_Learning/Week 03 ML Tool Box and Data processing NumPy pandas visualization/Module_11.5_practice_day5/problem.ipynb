#### 1 . Create a 1D, 2D, and 3D NumPy array and print their dimensions and shapes.


```python
import numpy as np

# 1D array
arr1 = np.array([1,2,3,4,5])

print("1D Dimention",arr1.ndim)
print("1D shape", arr1.shape)

#2D array
arr2 = np.array([[1,2,3],
                 [4,5,6]
                ])

print("2D Dimention",arr2.ndim)
print("2D shape", arr2.shape)

#3D array
arr3 = np.array([
    [
        [1,2,3],
        [4,5,6]
    ],
    [
        [1,2,3],
        [4,5,6]
    ],
    [
        [1,2,3],
        [4,5,6]
    ]
])

print("3D Dimention",arr3.ndim)
print("3D shape", arr3.shape)
```

#### 4 . Create a 4×4 matrix from 0–15. Then , extract:
#### the first row
#### the last column
#### a 2×2 subarray from the center

```python
matrix_arr = np.arange(16).reshape(4,4)
print("Matrix: \n", matrix_arr)

fst_row = matrix_arr[0]
print("First row: ",fst_row)

lst_col = matrix_arr[:, -1]
print("Last column:", lst_col)

mid = matrix_arr[1:3, 1:3]
print("Center 2x2\n",mid)
```

#### 5 . Add a 1D array [1,2,3] to each row of a 3×3 matrix.


```python
matrix = np.array([
    [10,20,30],
    [40,50,60],
    [70,80,90]
])
arr = np.array([1,2,3])
ans = matrix + arr
print(ans)
```

#### 6 . Given two arrays A and B, compute:
#### (A + B) / 2
#### A**2 + B**2


```python
arrA = np.array([1,2,3])

arrB = np.array([1,2,3])

fst_ans = (arrA + arrB)/2

scnd_ans = (arrA**2 + arrB**2)

print(fst_ans)
print(scnd_ans)
```

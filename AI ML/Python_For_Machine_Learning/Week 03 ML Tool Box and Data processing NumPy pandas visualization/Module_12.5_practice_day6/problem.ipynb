#### Create a Pandas Series from the list [10, 20, 30, 40, 50]. Print the Series, its index, and data type. Then create a DataFrame with columns "Name", "Age", and "City" containing at least 4 rows of data.


```python
import pandas as pd 

series = pd.Series([10, 20, 30, 40, 50])
print(series)
print(series.index)
print(series.dtype)

data = [
    {'Name': 'Tamim', 'Age': 20, 'City': 'Noakhali'},
    {'Name': 'Shajid', 'Age': 21, 'City': 'Feni'},
    {'Name': 'Emon', 'Age': 22, 'City': 'Miroshorai'},
    {'Name': 'Mannan', 'Age': 23, 'City': 'Noakhali'}
]
df = pd.DataFrame(data)
print(df)
```

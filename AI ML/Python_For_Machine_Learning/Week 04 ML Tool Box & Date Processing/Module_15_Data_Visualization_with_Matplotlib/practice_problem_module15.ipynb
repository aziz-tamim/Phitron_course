```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('student_IQdata.csv')
df
```

```python

plt.xlabel('Study Hours')
plt.ylabel('IQ Score')
plt.title('Study Hour and IQ Score Check')
plt.scatter(df['Study_Hour'], df['IQ_Score'])
plt.show()
```

```python
plt.xlabel('Shoe Size')
plt.ylabel('Study Hour')
plt.title('Shoe Size vs Study Hour')
plt.scatter(df['Shoe_Size'], df['Study_Hour'])
```

```python
plt.xlabel('Shoe Size')
plt.ylabel('Frequency')
plt.title('Shoe Size')
plt.hist(df['Shoe_Size'],color='Red', edgecolor='black')
```

## Answer To the Q. No 01

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

df = pd.read_csv('final-employee-ds.csv')
```

```python
convert_float = np.array(df['TrainingHours'], dtype = float)
print(np.mean(convert_float))
print(np.std(convert_float ))
```

### Answer To the Q. No 02

```python
df['Salary'] = df.groupby('Department')['Salary'].transform(lambda x : x.fillna(x.median()))

calculate_salary = df[(df['ExperienceYears'] > 12) & (df['ProjectCount'] > 5)]

salary_sum = calculate_salary['Salary'].sum()
print(salary_sum)
```

### Answer To the Q. No 03

```python
data_filter = df[((df['Department'] == 'IT') | (df['Department'] == 'Finance')) & (df['Age'] >= 30) & (df['Age'] <= 45) & (df['PerformanceScore'] > 88)]

desen = data_filter.sort_values('Salary', ascending = False)
output = desen[['Name', 'Department', 'Salary', 'PerformanceScore']]
print(output)
```

### Answer To the Q. No 04

```python
df['SalaryPerHour'] = df['Salary'] / (df['WorkHoursPerWeek'] * 4)

top5_employees = df.sort_values('SalaryPerHour', ascending=False).head(5)
final_output = top5_employees[['Name', 'SalaryPerHour']]
print(final_output)
```

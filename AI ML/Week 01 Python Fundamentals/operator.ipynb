## Operator in Python

```python
x = 10; y = 3
sum = x+y # summation
sub = x-y # subtraction
mul = x*y # multiplication
div = round(x/y, 2) # division
rem = x%y # remainder
print(sum, sub, mul, div, rem)
```

```python
x = 10; y = 3

greater_than = x>y
greater_than_or_equal = (10 >= 10)
print(greater_than, greater_than_or_equal)


less_than = x<y
less_than_or_equal = (10 <= 10)
print(less_than, less_than_or_equal)


equal = (x == y)
print(equal)


not_equal = (x != y)
print(not_equal)
```

## Logical Operator

```python
x = 10; y = 5; z = 15

result = (x>y) and (x>z)
print(result)
```

```python
x = 10; y = 5; z = 15

result = (x>y) or (x>z)
print(result)
```
